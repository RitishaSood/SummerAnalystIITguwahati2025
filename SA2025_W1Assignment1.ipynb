### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
# the median mpg
MedianMpg = df['mpg'].median()
# the cars that have mpg greater than the median mpg
AboveMedianMpg = df[df['mpg'] > MedianMpg].copy()
# the hp by weight ratio of this filtered dataSet
HpByWeight = AboveMedianMpg['horsepower'].astype(float) / AboveMedianMpg['weight'].astype(float)
# Car/Cars with best (max) hp by weight ratio
BestRatio = HpByWeight.max()
carNames = AboveMedianMpg[HpByWeight == BestRatio]['NAME']
print("The car/cars with Best Ratio of", BestRatio, "of HP by Weight and above median mpg is/are", list(carNames))
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
# Group by year and origin to calculate the mean
grouped = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()
# Plot using Seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(data=grouped, x='model_year', y='mpg', hue='origin', marker='o')
plt.title('Evolution of Average MPG Over Years by Origin')
plt.xlabel('model_year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.tight_layout()
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
#your code here 
# Clean the data: convert to correct types and drop missing values
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df['weight'] = pd.to_numeric(df['weight'], errors='coerce')
df['mpg'] = pd.to_numeric(df['mpg'], errors='coerce')
df['origin'] = df['origin'].astype(str).str.lower()
df = df.dropna(subset=['horsepower', 'weight', 'mpg', 'origin'])

# Filter only the specified origins
hue_order = ['japan', 'europe', 'usa']
df = df[df['origin'].isin(hue_order)]

# Create the scatterplot
plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=df,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    hue_order=hue_order,
    palette='Set2',
    sizes=(40, 400),
    alpha=0.7,
    edgecolor='k'
)

plt.title('Horsepower vs Weight: Colored by Origin, Sized by MPG', fontsize=16)
plt.xlabel('Horsepower', fontsize=14)
plt.ylabel('Weight', fontsize=14)
plt.legend(title='Origin', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
# Clean column names for consistency
df.columns = df.columns.str.lower().str.strip()
# Group by car name and count unique years
name_year_count = df.groupby('name')['model_year'].nunique()
# Filter car names appearing in more than one year
multi_year_names = name_year_count[name_year_count > 1].index
# Filter original DataFrame
df_multi = df[df['name'].isin(multi_year_names)]
# Group by name and calculate std, count, and mean for mpg
stats = df_multi.groupby('name')['mpg'].agg(['std', 'count', 'mean']).reset_index()
# Keep cars with at least 2 appearances and std(mpg) < 1.0
result = stats[(stats['count'] >= 2) & (stats['std'] < 1.0)]
# Sort by count descending, then mean mpg descending
result = result.sort_values(['count', 'mean'], ascending=[False, False])
# Show only required columns, rounded for clarity
result = result[['name', 'count', 'mean']].rename(columns={'count': 'appearances', 'mean': 'avgmpg'})
result['avgmpg'] = result['avgmpg'].round(2)

print(result)
```
